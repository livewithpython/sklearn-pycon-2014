## Reading in the data

```python
import numpy as np

data = np.loadtxt('data/web_traffic.tsv', delimiter='\t')

print data.shape
```

```python
print data[:10]
```

## Prepare the data

```python
x = data[:, 0]
y = data[:, 1]

import scipy as sp

print"Number of invalid entries:", sp.sum(sp.isnan(y))
```

```python
x = x[~np.isnan(y)]
y = y[~np.isnan(y)]
```

## Taking a Peek

## Selecting a Model

```python
train = x[::, np.newaxis]
target = y.copy()
```

```python
from sklearn.linear_model import LinearRegression
linear = LinearRegression()
linear.fit(train, target)

print "Linear Model Score : ", linear.score(train, target)
```

```python
## LinearRegression 
from sklearn.linear_model import LinearRegression
linear = LinearRegression()
train = x[::, np.newaxis]
target = y.copy()
linear.fit(train, target)

print "Linear Model : ", linear.score(train, target)

## RandomForest
from sklearn.ensemble import RandomForestRegressor
random = RandomForestRegressor()
train = x[::, np.newaxis]
target = y.copy()
random.fit(train, target)

print "RandomForest Model : ", random.score(train, target)

## Polynomial degree 2
from sklearn.linear_model import Ridge
import numpy as np

ridge = Ridge()
rtrain = np.vander(x, 4 + 1)
target = y.copy()

ridge.fit(rtrain, target)

print "RidgeRegression Model : ", ridge.score(rtrain, target)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6),)
plt.scatter(train, target, )
#Linear
plt.plot(train, linear.predict(train), linewidth=4)

#RandomForest
plt.plot(train, random.predict(train), linewidth=4)

#Ridge
plt.plot(train, ridge.predict(rtrain), linewidth=4)

plt.legend(["Linear", "RandomForest", "RidgeRegression"], loc="upper left")

plt.ylim(ymax=5000, ymin=500)
#plt.xlim(xmax=10000)

plt.show()
```

```python
## Look into the future
import scipy as sp
mx_raw=sp.linspace(0 * 7 * 24, 10 * 7 * 24, 100)
mx = mx_raw[::, np.newaxis]

import matplotlib.pyplot as plt
plt.figure(figsize=(10, 5))
plt.scatter(train, target, s=10)

#Linear
plt.plot(mx, linear.predict(mx), linewidth=4)

#RandomForest
plt.plot(mx, random.predict(mx), linewidth=4)

#Ridge
plt.plot(mx, ridge.predict(np.vander(mx_raw, 4 + 1)), linewidth=4)

plt.legend(["Linear", "RandomForest", "RidgeRegression"], loc="upper left")

plt.ylim(ymin=0, ymax=10000)
plt.xlim(xmin=-100, xmax=900)

plt.show()
```

## Cross validation

```python
from sklearn import cross_validation as cv

def split_data(X_data, y_data):
    """ Split the dataset in train and test """
    return cv.train_test_split(X_data, y_data, test_size=0.1, random_state=0)

X_train, X_test, y_train, y_test = split_data(train, target)
#Linear
linear.fit(X_train, y_train)
print "Linear Model Score : ", linear.score(X_test, y_test)

#Random
random.fit(X_train, y_train)
print "RandomForest Model Score : ", random.score(X_test, y_test)

#RidgeRegression
rtrain = np.vander(X_train[::, 0], 4 + 1)
rtest = np.vander(X_test[::, 0], 4 + 1)
ridge.fit(rtrain, y_train)
print "RidgeRegression : ", ridge.score(rtest, y_test)
```

## Solution

```python
from scipy.optimize import fsolve
import scipy as p

fbt = sp.poly1d(ridge.coef_)
print(fbt - 100000)
```

```python
reached_max = fsolve(fbt - 100000, 800) / (7 * 24)
print("100,000 hits/hour expected at week %f" % reached_max[0])
```
